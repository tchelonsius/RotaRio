Queremos somente as do RJ, logo:

```python
cisp_df = cisp_df[cisp_df['munic'] == 'Rio de Janeiro']
```

```python
cisp_df.fillna(0, inplace=True)
cisp_df.dtypes
```

Removeremos muitas colunas por ter dados inuteis, ou que saem do nosso escopo (como crimes patrimoniais por exemplo)

```python
colunas_remover = [
    'regiao','mes_ano', 'mcirc', 'risp', 'munic', 'fase',  # metadados inúteis
    'feminicidio', 'tentativa_feminicidio',          # muitos nulos
    'lesao_corp_morte', 'lesao_corp_dolosa',         #Não são crimes de violência armada (foco)
    'lesao_corp_culposa', 'hom_culposo',
    'tentat_hom', 'estupro',
    'roubo_transeunte', 'roubo_celular',             
    'roubo_em_coletivo', 'roubo_rua',               
    'roubo_veiculo', 'roubo_carga', 'roubo_comercio',
    'roubo_residencia', 'roubo_banco', 'roubo_cx_eletronico',
    'roubo_conducao_saque', 'roubo_apos_saque',
    'roubo_bicicleta', 'outros_roubos', 'total_roubos',
    'furto_veiculos', 'furto_transeunte', 'furto_coletivo',
    'furto_celular', 'furto_bicicleta', 'outros_furtos', 'total_furtos',
    'sequestro', 'extorsao', 'sequestro_relampago', 'estelionato',
    'apreensao_drogas', 'posse_drogas', 'trafico_drogas',
    'apreensao_drogas_sem_autor', 'recuperacao_veiculos',
    'apf', 'aaapai', 'cmp', 'cmba',
    'ameaca', 'pessoas_desaparecidas', 'encontro_ossada',
    'pol_militares_mortos_serv', 'pol_civis_mortos_serv'
]

cisp_df = cisp_df.drop(columns=colunas_remover)
```

```python
cisp_df.columns
```

```python
len(cisp_df), cisp_df['cisp'].nunique()
```

Cisps mais violentas

```python
letalidade_por_cisp = (
    cisp_df.groupby('cisp')['letalidade_violenta']
    .sum()
    .sort_values(ascending=False)
    .head(15)
)
letalidade_por_cisp
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
palette = sns.color_palette("rocket_r", n_colors=15)

sns.barplot(
    x=letalidade_por_cisp.values,
    y=letalidade_por_cisp.index.astype(str),
    hue=letalidade_por_cisp.index.astype(str),
    palette=palette,
    legend=False,
    ax=ax
)

for bar, value in zip(ax.patches, letalidade_por_cisp.values):
    ax.text(
        bar.get_width() + 30,
        bar.get_y() + bar.get_height() / 2,
        f'{value:,}',
        ha='left', va='center',
        fontsize=10, fontweight='bold', color='#333333'
    )

ax.set_title('Top 10 CISPs por Letalidade Violenta Acumulada', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Letalidade Violenta (acumulada)', fontsize=11)
ax.set_ylabel('CISP', fontsize=11)
ax.set_xlim(0, letalidade_por_cisp.values.max() * 1.15)

fig.patch.set_facecolor("#e0e0e0")
ax.set_facecolor("#f5f5f5")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
evolucao_anual = cisp_df.groupby('ano')['letalidade_violenta'].sum()
evolucao_anual
```

Distribuição temporal da letalidade violenta no RJ 

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(evolucao_anual.index, evolucao_anual.values, color='#7b2d8b', linewidth=2.5, marker='o', markersize=5)
ax.fill_between(evolucao_anual.index, evolucao_anual.values, alpha=0.15, color='#7b2d8b')

for x, y in zip(evolucao_anual.index, evolucao_anual.values):
    ax.text(x, y + 150, f'{y:,}', ha='center', va='bottom', fontsize=7.5, fontweight='bold', color='#333333')

ax.set_title('Evolução da Letalidade Violenta no Município do Rio de Janeiro (2003–2026)', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Ano', fontsize=11)
ax.set_ylabel('Letalidade Violenta', fontsize=11)
ax.set_xticks(evolucao_anual.index)
ax.set_xticklabels(evolucao_anual.index, rotation=45)
ax.set_ylim(0, evolucao_anual.values.max() * 1.15)

fig.patch.set_facecolor("#e0e0e0")
ax.set_facecolor("#f5f5f5")
sns.despine()
ax.set_xlim(2002.5, 2026.5)
plt.tight_layout()
plt.show()
```
